```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('wine.csv')
df.sample(5)
```

```python
df['Wine'].unique()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.drop('Wine' , axis = 1) , df['Wine'] , test_size  = 0.2 , random_state = 0)
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mutual = mutual_info_classif(X_train , y_train)
mutual
```

```python
mutual = pd.Series(mutual)
mutual.index = X_train.columns
mutual.sort_values(ascending = False)
```

```python
mutual.sort_values(ascending = False).plot(kind = "bar")
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
sel_five_cols = SelectKBest(mutual_info_classif, k=5)

sel_five_cols.fit(X_train, y_train)
```

```python
X_train.columns[sel_five_cols.get_support()]
```
